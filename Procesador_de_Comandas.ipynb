```
%%writefile food_package.py
from abc import ABC, abstractmethod

class FoodPackage (ABC):
  @abstractmethod
  def pack(self)  -> str:
      pass

  @abstractmethod
  def material(self) -> str:
      pass

  def describe(self):
      return f"Empaque: {self.pack()}, Material: {self.material()}"

class Wrapping(FoodPackage):
   def pack(self):
    return "Food wrapping"

   def material(self):
    return "Aluminium"

class Bottle(FoodPackage):
   def pack(self):
    return "Plastic bottle"

   def material(self):
    return "Plastic"

class Glass(FoodPackage):
   def pack(self):
    return "Glass cup"

   def material(self):
    return "Glass"

class Box(FoodPackage):
   def pack(self):
    return "Meal Box"

   def material(self):
    return "Cardboard"
"""
print(Wrapping().describe())
print(Bottle().describe())
print(Glass().describe())
print(Box().describe())
"""
```

```
%%writefile product.py
from abc import ABC, abstractmethod
import food_package

class Product(ABC):
    def __init__(self,id:str,name:str,price:float):
      self.id = id
      self.name = name
      self.price = price

    def describe(self):
        return f"Product - Type: {self.type()}, Name: {self.name}, Id: {self.id} , Price: {self.price} , {self.foodPackage().describe()}."

    @abstractmethod
    def type(self) -> str:
        pass
    @abstractmethod
    def foodPackage(self)->food_package.FoodPackage:
        pass

class Hamburger(Product):
    def __init__(self,id:str,name:str,price:float):
        super().__init__(id,name,price)

    def type(self) -> str:
        return "Hamburguesa"

    def foodPackage(self):
        return food_package.Wrapping()

class Soda(Product):
    def __init__(self,id:str,name:str,price:float):
        super().__init__(id,name,price)

    def type(self) -> str:
        return "Soda"

    def foodPackage(self):
        return food_package.Bottle()

class Drink(Product):
    def __init__(self,id:str,name:str,price:float):
        super().__init__(id,name,price)

    def type(self) -> str:
        return "Bebida"

    def foodPackage(self):
        return food_package.Glass()

class HappyMeal(Product):
    def __init__(self,id:str,name:str,price:float):
        super().__init__(id,name,price)

    def type(self) -> str:
        return "Happy Meal"

    def foodPackage(self):
        return food_package.Box()
"""
print(Hamburger("swdrt", "Hamburguesa de queso", 1.20).describe())
print(Soda("vnjreln", "Soda light", 1.50).describe())
print(Drink("GTnjT", "Sprite de manzana", 3.45).describe())
print(HappyMeal("GHJbjbU", "Finguers de pollo con Ketchup", 4.05).describe())
"""
```

```
%%writefile user.py
from abc import ABC, abstractmethod

class User(ABC):
  def __init__(self,dni:str,name:str,age:int):
    self.dni = dni
    self.name = name
    self.age = age

  @abstractmethod
  def describe(self):
      pass

class Cashier(User):
  def __init__(self,dni:str,name:str,age:int,timeTable:str,salary:float):
    super().__init__(dni, name, age)
    self.timeTable = timeTable
    self.salary = salary

  def describe(self):
        return f"Cashier - Name: {self.name}, DNI: {self.dni} , Timetable: {self.timeTable}, Salary: {self.salary}."

class Customer(User):
  def __init__(self,dni:str,name:str,age:int,email:str,postalCode:str):
    super().__init__(dni, name, age)
    self.email = email
    self.postalCode = postalCode

  def describe(self):
        return f"Customer - Name: {self.name}, DNI: {self.dni} , Age: {self.age}, Email: {self.email}, Postal Code: {self.postalCode}"
"""
print(Cashier("215653", "Laura", 23, "Tardes", 25000).describe())
print(Cashier("78368", "Pepe", 38, "Mañanas", 30000).describe())
print(Customer("647389", "Luis", 17, "luisbreva17@gmail.com", "12002").describe())
"""
```

```
%%writefile converter.py
from abc import ABC, abstractmethod
import pandas as pd
import product
import user
import file_manager

class Converter(ABC):
  @abstractmethod
  def convert(self,dataFrame:pd.DataFrame) -> list:
      pass
  @staticmethod
  def informacion(listaObjetos:list):
    for obj in listaObjetos:
      print(obj.describe())

class CashierConverter(Converter):
  def convert(self,dataFrame:pd.DataFrame):
    cashiers_list = []
    for index, row in dataFrame.iterrows():
      cashier = user.Cashier(row['dni'], row['name'], row['age'], row['timetable'], row['salary'])
      cashiers_list.append(cashier)

    return cashiers_list

class CustomerConverter(Converter):
  def convert(self,dataFrame:pd.DataFrame):
    customers_list = []
    for index, row in dataFrame.iterrows():
      customer = user.Customer(row['dni'], row['name'], row['age'], row['email'], row['postalcode'])
      customers_list.append(customer)

    return customers_list

class ProductConverter(Converter):
  def convert(self,dataFrame:pd.DataFrame):
    products_list = []
    for index, row in dataFrame.iterrows():
      if "H" in row["id"] and "M" not in row["id"]:
        product_item = product.Hamburger(row['id'], row['name'], row['price'])
      elif "G" in row["id"]:
        product_item = product.Soda(row['id'], row['name'], row['price'])
      elif "B" in row["id"]:
        product_item = product.Drink(row['id'], row['name'], row['price'])
      elif "HM" in row["id"]:
        product_item = product.HappyMeal(row['id'], row['name'], row['price'])

      products_list.append(product_item)

    return products_list

"""
dfCajeros = file_manager.CSVFileManager().read("/content/cashiers.csv")
dfClientes = file_manager.CSVFileManager().read("/content/customers.csv")
dfHamburguesas = file_manager.CSVFileManager().read("/content/hamburgers.csv")
dfSodas = file_manager.CSVFileManager().read("/content/sodas.csv")
dfBebidas = file_manager.CSVFileManager().read("/content/drinks.csv")
dfHappyMeals = file_manager.CSVFileManager().read("/content/happyMeal.csv")

listaCajeros = CashierConverter().convert(dfCajeros)
listaClientes = CustomerConverter().convert(dfClientes)
listaHamburguesas = ProductConverter().convert(dfHamburguesas)
listaSodas = ProductConverter().convert(dfSodas)
listaBebidas = ProductConverter().convert(dfBebidas)
listaHappyMeals = ProductConverter().convert(dfHappyMeals)

CashierConverter().informacion(listaCajeros)
print("\n")
CustomerConverter().informacion(listaClientes)
print("\n")
ProductConverter().informacion(listaHamburguesas)
print("\n")
ProductConverter().informacion(listaSodas)
print("\n")
ProductConverter().informacion(listaBebidas)
print("\n")
ProductConverter().informacion(listaHappyMeals)
print("\n")
"""
```

```
%%writefile order.py
from user import User
from product import Product

class Order:
  def __init__(self, cashier:User, customer:User):
    self.cashier = cashier
    self.customer = customer
    self.lista = []

  def add(self, product:Product):
    self.lista.append(product)
    print(f"Producto '{product.name}' añadido a la orden.")

  def calculoTotal(self):
    total = 0
    for obj in self.lista:
      total = total + obj.price

    return round(total, 2)

  def show(self):
    print("The:\n" + self.customer.describe() + "\nWas attended by:\n" + self.cashier.describe())
    print("PRODUCTS ORDEREDD:")
    for obj in self.lista:
      print(obj.describe())
    print(f"Total price: {self.calculateTotal()}")
```
